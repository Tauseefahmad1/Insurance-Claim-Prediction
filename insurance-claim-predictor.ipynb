```python
#importing all the relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns

from sklearn import model_selection as mdsel
from sklearn.feature_selection import VarianceThreshold
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
#loading the data set
ins=pd.read_csv('train_dataset.csv')
ins.shape
```

```python
#let's get a look at our dataset
ins.head()
```

```python
#let's begin data exploration
ins.describe()
```

```python
ins.info()
```

```python
#now we know that we don't have any null values in any of the fields
#we should check variance of all the features so we can use only high variance features as low variance features don't affect target anyways
ins_features=ins.drop('is_claim',axis=1)
ins_var=VarianceThreshold(threshold=0.2)
ins_var.fit_transform(ins_features)
for feature in zip(ins_features.columns,ins_var.variances_):
    print(feature)
```

```python
#hence, we use only those features which have variance greater than 0.2
not_features=['policy_tenure','age_of_car','age_of_policyholder','is_esc','is_tpms','is_parking_sensors','rear_brakes_type','gear_box','turning_radius','is_rear_window_wiper','is_rear_window_washer','is_power_steering','is_speed_alert']
ins_features=ins_features.drop(not_features,axis=1)
ins_features.shape
```

```python
#taking the target variable as a separate column
ins_target=ins['is_claim']
ins_target.shape
```

```python
#splitting given data into train and test sections
X_train,X_test,y_train,y_test=mdsel.train_test_split(ins_features,ins_target,test_size=0.35)
X_train.shape
```

```python
X_train.head()
```

```python
#since all the features columns are more or less independent of each other, we can use a naive bayes classifier
#hence importing the model
from sklearn.naive_bayes import GaussianNB
nbc=GaussianNB()
nbc.fit(X_train,y_train)
```

```python
#making predictions on the test dataset
y_pred=nbc.predict(X_test)
y_pred
```

```python
#checking the accuracy of our model
print(accuracy_score(y_pred,y_test)*100)
```

```python
#now let's import the test data..for which our model will be evaluated
actual_test=pd.read_csv('test_dataset.csv')
actual_test.head()
```

```python
#dropping useless columns
actual_features=actual_test.drop(not_features,axis=1)
actual_features.head()
```

```python
#making predictions
actual_predictions=nbc.predict(actual_features)
actual_predictions
```

```python
#converting the array to an appropriate dataframe
predictions_df=pd.DataFrame(actual_predictions)
predictions_df.index=actual_test.index
predictions_df.columns=['is_claim']
predictions_df.head()
```

```python
#saving the predictions dataframe as a csv file to be uploaded
predictions_df.to_csv("prediction_results.csv", index = False)
```
